For more info on the Boundaries: https://www2.minneapolismn.gov/government/departments/coordinator/sustainability/policies/green-zones-initiative/green-zones-common-questions/

```python
# Import libraries

import geopandas as gpd # Spatial Data manipulation
import matplotlib.pyplot as plt # Basic Plotting
import contextily # Basemaps
```

```python
# Load data

# GZ Boundary
gz_bounds = gpd.read_file('gz_boundaryUTM15N.geojson')

# MPLS Boundary

path = '../Boundary/mpls_boundary.geojson'
mpls_bounds = gpd.read_file(path)
```

```python
gz_bounds.head()
```

```python
# Plotting

f, ax = plt.subplots(1,1,figsize = (8,8))

mpls_bounds.plot(ax = ax,
                cmap = 'cool',
                alpha = 0.3)

gz_bounds.plot('Id',
               ax = ax,
               cmap = 'Greens_r',
               alpha = 0.4)

contextily.add_basemap(ax = ax, 
                       crs = mpls_bounds.crs, 
                       source = contextily.providers.Stamen.TonerLite
                      )
```
